```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def clamp(x, min_value, max_value):
    return max(min(x, max_value), min_value)

# Generate data
x = np.linspace(-10, 10, 400)
y = [clamp(val, -5, 5) for val in x]

# Plot the clamp function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='clamp(x, -5, 5)')
plt.xlabel('x')
plt.ylabel('clamp(x)')
plt.title('Clamp Function Visualization')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Visualize the smooth step function
def smooth_step(x, edge0, edge1):
    # Scale, bias and saturate x to 0..1 range
    x = clamp((x - edge0) / (edge1 - edge0), 0.0, 1.0)
    # Evaluate polynomial
    return x * x * (3 - 2 * x)

# Generate data
x = np.linspace(-10, 10, 400)
y1 = [smooth_step(val, -5, 5) for val in x]
y2 = [smooth_step(val, 2, -2) for val in x]

# Plot the smooth step function
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='smooth_step(x, -5, 5)')
plt.plot(x, y2, label='smooth_step(x, 2, -2)')
plt.xlabel('x')
plt.ylabel('smooth_step(x)')
plt.title('Smooth Step Function Visualization')
plt.legend()
plt.grid(True)
plt.show()

```
